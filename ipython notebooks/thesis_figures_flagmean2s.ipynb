```python
import numpy as np
import scipy as sp
import distances
import gr_kmeans
import gr_lbg
import heatmap
import plots
import matplotlib.pyplot as plt
from flag_mean import flag_mean
from mpl_toolkits import mplot3d
from load_ip_data import load_ip_data
from mnist import load_mnist_data
from subspaces import make_subspaces
```

```python
data, labels = load_mnist_data(2, sample_size = 1000)
print(data.shape)
```

```python
sub, lbls = make_subspaces(data, labels, dim=5)
print(sub[0].shape)
print(len(sub))
```

```python
center = flag_mean(sub)
```

```python
all_pts = sub + [center]
D = distances.chordal_distance(all_pts, all_pts)
embed = distances.mds(D, eigplot=True)
```

```python
plt.figure()
for i in np.unique(lbls):
    idx = np.where(lbls == i)[0]
    plt.plot(embed[:, 0], embed[:, 1], 'o', label='Data')
plt.plot(embed[-1,0], embed[-1,1], 'o', markeredgecolor='k', markersize=8, label='Center')
plt.legend()
plt.title('MDS Embedding of MNIST points in Gr(5,784)')
plt.show()
```
