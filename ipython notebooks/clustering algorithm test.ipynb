```python
import numpy as np
import scipy as sp
import distances
import gr_kmeans
import gr_lbg
import matplotlib.pyplot as plt
```

```python
def generate_point_clouds(m,n, clusters = 2, cluster_size = 10, eps=.1):
    point_clouds = []
    labels = []
    for c in range(clusters):
        point_clouds.append(np.linalg.qr(np.random.random((m,n))*2-1)[0])
        labels.append(c)
        for i in range(cluster_size-1):
            point_clouds.append(np.linalg.qr(point_clouds[c*cluster_size]+eps*np.random.random((m,n))*2-1)[0])
            labels.append(c)
    return point_clouds, labels
    
```

```python
data, labels = generate_point_clouds(10,2,clusters=3)
D = distances.chordal_distance(data,data)
embed = distances.mds(D)
plt.figure()
for i in np.unique(labels):
    idx = np.where(labels == i)[0]
    plt.plot(embed[idx,0], embed[idx,1], 'o')
plt.show()
```

```python
grlbg = gr_lbg.gr_lbg()
grlbg.fit(data, center_count=3, plot_results = True, eigplot=True)
print(grlbg.distortion_change)
```
