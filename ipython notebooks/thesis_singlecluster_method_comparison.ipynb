```python
import numpy as np
import scipy as sp
import distances
import gr_kmeans
import gr_lbg
import plots
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from load_ip_data import load_ip_data
from mnist import load_mnist_data
from subspaces import make_subspaces
```

```python
def generate_point_clouds(m,n, clusters = 1, cluster_size = 10, eps=.1):
    point_clouds = []
    labels = []
    for c in range(clusters):
        point_clouds.append(np.linalg.qr(np.random.random((m,n))*2-1)[0])
        labels.append(c)
        for i in range(cluster_size-1):
            point_clouds.append(np.linalg.qr(point_clouds[c*cluster_size]+eps*(np.random.random((m,n))*2-1))[0])
            labels.append(c)
    return point_clouds, labels
```

```python
cluster_size = 100
clusters=1
data, labels = generate_point_clouds(10,2, clusters=clusters, cluster_size=cluster_size, eps=.1)
```

```python
kmeans = gr_kmeans.gr_kmeans()
kmeans_centers, kmeans_labels = kmeans.fit(data, true_labels=labels, center_count=1, plot_results=True, eigplot=True,
                                          show_cluster_data=False, numits=1)
```

```python
lbg = gr_lbg.gr_lbg()
lbg_centers, lbg_labels = lbg.fit(data, true_labels=labels, center_count=1, plot_results=True, eigplot=True,
                                 show_cluster_data=False, numits=100)
```

```python
print(distances.chordal_distance(lbg_centers, kmeans_centers))
```
